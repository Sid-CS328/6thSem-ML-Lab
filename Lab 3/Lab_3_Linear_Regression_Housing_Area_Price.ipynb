```python
# Using Library
# Expected Prediction: ≈ $41288


import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression


df = pd.read_csv("canada_per_capita_income.csv")


X = df[['year']]
y = df['per capita income (US$)']


model = LinearRegression()
model.fit(X, y)


prediction_2020 = model.predict(pd.DataFrame([[2020]], columns=['year']))

print("Predicted per capita income in 2020:", prediction_2020[0])

```

```python
#Without using libraries

import numpy as np
import pandas as pd

df = pd.read_csv("canada_per_capita_income.csv")

X = df['year'].values
y = df['per capita income (US$)'].values


X_b = np.c_[np.ones(len(X)), X]


beta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

intercept = beta[0]
slope = beta[1]


prediction_2020 = intercept + slope * 2020

print("Intercept:", intercept)
print("Slope:", slope)
print("Predicted income in 2020:", prediction_2020)
```

```python
# Using library
# Expected Prediction: ≈ $139000


import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("salary.csv")

df = df.dropna()

X = df[['YearsExperience']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)

prediction = model.predict(pd.DataFrame([[12]], columns=['YearsExperience']))
print("Predicted salary for 12 years experience:", prediction[0])
```

```python
# Without library

import numpy as np
import pandas as pd

df = pd.read_csv("salary.csv")
df = df.dropna()

X = df['YearsExperience'].values
y = df['Salary'].values


X_b = np.c_[np.ones(len(X)), X]


beta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

intercept = beta[0]
slope = beta[1]


prediction = intercept + slope * 12

print("Intercept:", intercept)
print("Slope:", slope)
print("Predicted Salary:", prediction)
```
