```python
# Handling Missing Values

adult_df.replace('?', np.nan, inplace=True)

imputer = SimpleImputer(strategy='most_frequent')
adult_df = pd.DataFrame(imputer.fit_transform(adult_df),
                        columns=adult_df.columns)
```

```python
# Handling Categorical Data

categorical_cols = adult_df.select_dtypes(include='object').columns

encoder = LabelEncoder()
for col in categorical_cols:
    adult_df[col] = encoder.fit_transform(adult_df[col])
```

```python
# Handling Outliers

numeric_cols = adult_df.select_dtypes(include=np.number).columns

Q1 = adult_df[numeric_cols].quantile(0.25)
Q3 = adult_df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

adult_no_outliers = adult_df[
    ~((adult_df[numeric_cols] < (Q1 - 1.5 * IQR)) |
      (adult_df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)
]

print("Original shape:", adult_df.shape)
print("After outlier removal:", adult_no_outliers.shape)
```

```python
# Min-Max Scaling (Normalization)

minmax_scaler = MinMaxScaler()

adult_minmax = adult_no_outliers.copy()
adult_minmax[numeric_cols] = minmax_scaler.fit_transform(
    adult_no_outliers[numeric_cols]
)
```

```python
# Standard Scaling

standard_scaler = StandardScaler()

adult_standard = adult_no_outliers.copy()
adult_standard[numeric_cols] = standard_scaler.fit_transform(
    adult_no_outliers[numeric_cols]
)
```
